```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dfset into a Pandas dfframe
df = pd.read_csv('data.csv')
```

```python
# Print the first 5 rows of the dfframe
df.head()

# Print the last 5 rows of the dfframe
df.tail()

# Check the shape of the dfframe
df.shape

# Check the df types of the columns
df.dtypes

# Summary statistics of the numerical columns
df.describe()
```

```python
# Clean the df if necessary
df.isnull().sum()  ## count the number of missing values in each column
df.dropna()  ## remove rows with missing values
df.fillna(0)  ## replace missing values with 0
df.duplicated().sum()  ## count the number of duplicate rows
df.drop_duplicates()  ## remove duplicate rows
```

```python
# Create a correlation matrix and plot it as a heatmap
corr_matrix = df.iloc[:,2:32].corr()
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.show()
```

```python
# Plot a histogram of a numerical column
plt.hist(df['symmetry_worst'], bins=20)
plt.xlabel('symmetry_worst')
plt.ylabel('Frequency')
plt.title('Histogram of symmetry_worst')
plt.show()
```

```python
# Explore the distribution of all numeric variables
# Create subplots for each column
fig, axes = plt.subplots(nrows=5, ncols=6, figsize=(10,6))

# Loop through the columns and plot each column in a subplot
for i, ax in enumerate(axes.flatten()):
    if i < len(df.iloc[:, 2:32].columns):
        col = df.iloc[:, 2:32].columns[i]
        df[[col]].plot(ax=ax, kind="hist")
        
# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Boxplot showing a variable between the target groups
sns.boxplot(x='diagnosis', y='area_worst', data=df)
plt.show()
```

```python
# Dimensional reduction
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Scale the data
data_scaled = StandardScaler().fit_transform(df.iloc[:,2:32])

# Run PCA
pca = PCA()
pca.fit(data_scaled)
pca_data = pca.transform(data_scaled)

# Create a scree plot to visualize the explained variance of each principal component
plt.clf()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.show()

# Plot the first two principal components
plt.clf()
plt.scatter(pca_data[:, 0], pca_data[:, 1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
